```python
import finterstellar as fs
```

```python
df = fs.get_price('139480.KS', start_date='2021-01-01', end_date='2021-12-02')
fs.rsi(df, w=14)
#fs.draw_chart(df, left='rsi', right='SPYD')
fs.draw_chart(df, right='139480.KS')
```

```python
fs.indicator_to_signal(df, factor='rsi', buy=40, sell=60)
fs.position(df)
fs.draw_chart(df, left='rsi', right='position_chart')
fs.evaluate(df, cost=.001)
fs.performance(df, rf_rate=0.01)
```
